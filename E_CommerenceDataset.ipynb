```python
import pandas as pd

df = pd.read_csv("ecommerce_dataset_updated.csv")

df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
df = df.dropna()  # or fill depending on column
```

```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.replace(r"[()%\.]", "", regex=True)

print(df.columns)
```

```python
df['revenue'] = df['final_pricers']
```

```python
df['discount_value'] = df['price_rs'] - df['final_pricers']
```

```python
df['purchase_date'] = pd.to_datetime(df['purchase_date'], dayfirst=True)
```

```python
df['month'] = df['purchase_date'].dt.month
df['year'] = df['purchase_date'].dt.year
```
